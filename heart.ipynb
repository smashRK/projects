# outliers boxplot

```python
fig = plt.figure(figsize=[16,12])
fig.suptitle('BOXPLOT OF dATA', fontsize=18, fontweight='bold')
fig.subplots_adjust(top=0.92);
fig.subplots_adjust(hspace=0.5, wspace=0.4);
for i ,col in enumerate(new_col):
    
    ax1 = fig.add_subplot(3, 2, i+1);
    ax1 = sns.boxplot(data = df, x=col);
    
    ax1.set_title(f'{col}', fontsize = 18)
```

```python
dict = {}
for col in ['RestingBP','Cholesterol','MaxHR','Oldpeak']:
    percentile25 = df[col].quantile(0.25)
    percentile75 = df[col].quantile(0.75)
    IQR = percentile75 - percentile25
    upper_limit = percentile75 + 1.5 * IQR
    lower_limit = percentile25 - 1.5 * IQR
    dict['upper_limit'+ '_' + col] = upper_limit
    dict['lower_limit'+ '_' + col] = lower_limit
```

```python
dict
```

```python
for col in ['RestingBP','Cholesterol','MaxHR','Oldpeak']:
    print('There are total {0} Customers data which {1} are less than lower limit.'.format((df[col]<dict['upper_limit'+ '_' + col]).sum(),col))
    print('There are total {} Customers data which {} are more than upper limit.'.format((df[col]>dict['upper_limit'+ '_' + col]).sum(),col))
```

# removing outliers

```python
for col in ['RestingBP','Cholesterol','MaxHR','Oldpeak']:

    df[col] = np.where(
        df[col] > dict['upper_limit_' + col],
        dict['upper_limit_' + col],
            np.where(
            df[col] < dict['lower_limit_' + col],
            dict['lower_limit_' + col],
            df[col] 
        )
    )
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split( x, y, test_size=0.33, random_state=55)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import recall_score , classification_report , confusion_matrix, accuracy_score, precision_score
from sklearn.metrics import precision_recall_curve , auc ,f1_score , plot_confusion_matrix
```

```python

model_list = []
accuracy_list = []
recall_list = []
precision_list = []
f1_score_list= [] 
```

```python
def Model_features(x_train , y_train , x_test , y_test , y_pred , classifier , model_name):

    accuracy ,precision , recall , f1_s = round(accuracy_score(y_test , y_pred),ndigits=3), round(precision_score(y_test,y_pred),ndigits=3) , round(recall_score(y_test,y_pred),ndigits=3), round(f1_score(y_test,y_pred),ndigits=3)
    
    print(f'Accuracy Score is :{accuracy}')
    print(f'Precision Score is :{precision}')
    print(f'Recall Score is :{recall}')
    print(f'f1 Score is :{f1_s}')
    model_list.append(model_name)
    accuracy_list.append(accuracy)
    recall_list.append(recall)
    precision_list.append(precision)
    f1_score_list.append(f1_s)
 

    print(metrics.classification_report(y_test, y_pred))
    
```

```python
def confusion_matrix_plot(y_test ,y_pred, classifier ,classifier_name):
    sns.heatmap(confusion_matrix(y_test ,y_pred), annot=True)
```

##logistic regression

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
lr.score(x_test,y_test)
```

```python
Model_features(x_train , y_train , x_test , y_test , y_pred , lr , "LogisticRegression" )
confusion_matrix_plot(y_test ,y_pred , lr , "LogisticRegression")
```

# Knn classifier

```python
knn = KNeighborsClassifier(n_neighbors=20,algorithm='ball_tree')
knn.fit(x_train,y_train)
y_pred= knn.predict(x_test)
knn.score(x_test,y_test)
```

```python
Model_features(x_train , y_train , x_test , y_test , y_pred , knn , "KNeighborsClassifier" )
confusion_matrix_plot(y_test ,y_pred , knn, "KNeighborsClassifier")
```

# randon forest

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
```

```python
Model_features(x_train , y_train , x_test , y_test , y_pred , rf , "RandomForestClassifier" )
confusion_matrix_plot(y_test ,y_pred , rf, "RandomForestClassifier")
```

# decision tree

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
```

```python
Model_features(x_train , y_train , x_test , y_test , y_pred , dt , "DecisionTreeClassifier" )
confusion_matrix_plot(y_test ,y_pred , dt, "DecisionTreeClassifier")
```
