```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:/downloads/insurance.csv")
```

```python
df.head()
```

```python
sex={'male':0,'female':1}
smoker={'yes':0,'no':1}
region={'northeast':0,'northwest':1,'southeast':2,'southwest':3}
df.sex=[sex[i] for i in df.sex]
df.smoker=[smoker[j] for j in df.smoker]
df.region=[region[k] for k in df.region]
```

```python
df
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
ml = LinearRegression()
ml.fit(x_train,y_train)
```

```python
print(ml.intercept_)
```

```python
print(ml.coef_)
```

```python
y_pred=ml.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
print('mean absolute error',metrics.mean_absolute_error(y_pred,y_test))
print('mean squared error',metrics.mean_squared_error(y_pred,y_test))
print('root mean sqared error',np.sqrt(metrics.mean_squared_error(y_pred,y_test)))
```
