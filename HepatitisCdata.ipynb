# Outliers 

```python
out_col = ['Age','ALB','ALP','ALT','AST','BIL','CHE','CHOL','CREA','GGT','PROT']
```

```python
df_num = df[out_col]
df_num.describe()

Q1 = df_num.quantile(0.25)

Q3 = df_num.quantile(0.75)

IQR = Q3 - Q1

IQR
((df_num < (Q1 - 1.5*IQR)) | (df_num > (Q3 + 1.5 *IQR))).any()
```

```python
dict = {}
for i,col in enumerate(out_col):
    percentile25 = df[col].quantile(0.25)
    percentile75 = df[col].quantile(0.75)
    IQR = percentile75 - percentile25
    upper_limit = percentile75 + 1.5 * IQR
    lower_limit = percentile25 - 1.5 * IQR
    dict['upper_limit'+ '_' + col] = upper_limit
    dict['lower_limit'+ '_' + col] = lower_limit
```

```python
dict
```

```python
for i,col in enumerate(out_col):
    print('There are total {0} Customers data which {1} are less than lower limit.'.format((df[col]<dict['upper_limit'+ '_' + col]).sum(),col))
    print('There are total {} Customers data which {} are more than upper limit.'.format((df[col]>dict['upper_limit'+ '_' + col]).sum(),col))
```

```python
for i,col in enumerate(out_col):

    df[col] = np.where(df[col] > dict['upper_limit_' + col],dict['upper_limit_' + col],
            np.where(df[col] < dict['lower_limit_' + col],dict['lower_limit_' + col],
            df[col]))
```

```python
df_num = df[out_col]
df_num.describe()

Q1 = df_num.quantile(0.25)

Q3 = df_num.quantile(0.75)

IQR = Q3 - Q1

IQR
((df_num < (Q1 - 1.5*IQR)) | (df_num > (Q3 + 1.5 *IQR))).any()
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split( x, y, test_size=0.33, random_state=55)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import recall_score , classification_report , confusion_matrix, accuracy_score, precision_score
from sklearn.metrics import f1_score , ConfusionMatrixDisplay
import warnings
warnings.filterwarnings('ignore')
```

```python

model_list = []
accuracy_list = []
recall_list = []
precision_list = []
f1_score_list= [] 
```

```python
def Model_features(x_train , y_train , x_test , y_test , y_pred , classifier , model_name):

    accuracy ,precision , recall , f1_s = round(accuracy_score(y_test , y_pred),ndigits=2), round(precision_score(y_test,y_pred,average='weighted'),ndigits=2) , round(recall_score(y_test,y_pred,average='weighted'),ndigits=2), round(f1_score(y_test,y_pred,average='weighted'),ndigits=2)
    
    print(f'Accuracy Score is :{accuracy}')
    print(f'Precision Score is :{precision}')
    print(f'Recall Score is :{recall}')
    print(f'f1 Score is :{f1_s}')
    model_list.append(model_name)
    accuracy_list.append(accuracy)
    recall_list.append(recall)
    precision_list.append(precision)
    f1_score_list.append(f1_s)
 

    print(metrics.classification_report(y_test, y_pred))
```

```python
def confusion_matrix_plot( y_test ,y_pred, classifier ,classifier_name):
    sns.heatmap(confusion_matrix(y_test ,y_pred), annot=True)
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
Model_features(x_train , y_train , x_test , y_test , y_pred , lr , "LogisticRegression" )
confusion_matrix_plot(y_test , y_pred , lr , "LogisticRegression")
```

# knn


```python
knn = KNeighborsClassifier(n_neighbors=20,algorithm='ball_tree')
knn.fit(x_train,y_train)
y_pred= knn.predict(x_test)
```

```python
Model_features(x_train , y_train , x_test , y_test , y_pred , knn , "KNeighborsClassifier" )
confusion_matrix_plot(y_test, y_pred , knn, "KNeighborsClassifier")
```

# Random forest

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
```

```python
Model_features(x_train , y_train , x_test , y_test , y_pred , rf , "RandomForestClassifier" )
confusion_matrix_plot(y_test, y_pred , rf, "RandomForestClassifier")
```
